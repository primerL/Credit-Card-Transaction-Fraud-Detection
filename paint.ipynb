```python
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False 

# Define different methods and their corresponding F1 scores
methods = ["Decision Tree", "Random Forest", "AdaBoost"]
f1_scores_original = [0.783625730994152, 0.6713286713286712, 0.7904191616766466]  # Original data
f1_scores_undersample = [0.1275109170305677, 0.22652757078986588, 0.05298457411133467]  # Random undersampling
f1_scores_tomek = [0.7976878612716763, 0.6713286713286712, 0.7904191616766466]  # Tomek Links
f1_scores_original_intel = [0.783625730994152, 0.6805555555555556, 0.7904191616766466]  # Original data Intel
f1_scores_undersample_intel = [0.1275109170305677, 0.19240506329113924, 0.05298457411133467]  # Random undersampling Intel
f1_scores_tomek_intel = [0.7976878612716763, 0.6805555555555556, 0.7904191616766466]  

# Set the width of the bar chart
bar_width = 0.15

# Generate x-axis coordinates
x = range(len(methods))

# Create a bar chart
plt.figure(figsize=(12, 6))

# Plot the F1 scores for different methods and add text labels
bar_original = plt.bar(x, f1_scores_original, bar_width, label="Original Data", color='b')
bar_undersample = plt.bar([i + bar_width for i in x], f1_scores_undersample, bar_width, label="Random Undersampling", color='g')
bar_tomek = plt.bar([i + bar_width * 2 for i in x], f1_scores_tomek, bar_width, label="Tomek Links", color='r')
bar_original_intel = plt.bar([i + bar_width * 3 for i in x], f1_scores_original_intel, bar_width, label="Original Data Intel", color='c')
bar_undersample_intel = plt.bar([i + bar_width * 4 for i in x], f1_scores_undersample_intel, bar_width, label="Random Undersampling Intel", color='m')
bar_tomek_intel = plt.bar([i + bar_width * 5 for i in x], f1_scores_tomek_intel, bar_width, label="Tomek Links Intel", color='y')

# Set x-axis labels
plt.xticks([i + bar_width * 2 for i in x], methods)

plt.xlabel("Method")
plt.ylabel("F1 Score")
plt.title("Performance Comparison of Different Methods")

# Add text labels on top of each bar
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, '{:.4f}'.format(height), ha='center', va='bottom')

add_labels(bar_original)
add_labels(bar_undersample)
add_labels(bar_tomek)
add_labels(bar_original_intel)
add_labels(bar_undersample_intel)
add_labels(bar_tomek_intel)

# Add a legend
plt.legend()

# Show the chart
plt.show()
```
